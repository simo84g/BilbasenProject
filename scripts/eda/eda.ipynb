```python
#Check mising values again
print(df.isnull().sum())
```

When looking into these missing values, i can see that there are many values with around 46-47 missing values. 
Maybe that means that 46 observations are missing data in all variables

```python
# Count the number of missing values per row
missing_per_row = df.isnull().sum(axis=1)

# Add this information to the DataFrame for analysis
df['Missing Count'] = missing_per_row

# Optionally, view a summary of rows grouped by their missing counts
print(df['Missing Count'].value_counts())


# Create a DataFrame containing only rows with missing values
rows_with_missing = df[df['Missing Count'] > 0]

# Inspect the first few rows to see a pattern
print(rows_with_missing)
```

```python
# Count the number of missing values per row
df['Missing Count'] = df.isnull().sum(axis=1)

# Filter rows that have 28 or more missing values
rows_with_28_or_more_missing = df[df['Missing Count'] >= 28]

# Display these observations
print(rows_with_28_or_more_missing)
```

```python
df.shape
```

```python
# Filter rows that have 28 or more missing values
rows_to_delete = df[df['Missing Count'] >= 28].index

# Drop the rows with 28 or more missing values
df.drop(rows_to_delete, inplace=True)

# Drop the 'Missing Count' column as it's no longer needed
df.drop('Missing Count', axis=1, inplace=True)

# Confirm the removal by displaying the shape of the DataFrame
print("Shape after removing rows:", df.shape)
```

```python
#Check mising values again
print(df.isnull().sum())
```

Mofification of variables

```python
df_v['Kilometertal'] = df_v['Kilometertal'].fillna("0") #All Nan to 0
df_v['Kilometertal'] = df_v['Kilometertal'].str.replace('km', '', regex=False).str.strip() #Remove 'Km.' from the kilometertal
df_v['Kilometertal'] = df_v['Kilometertal'].replace(r'.*\(ny bil\).*', '0', regex=True)
df_v['Trækhjul'] = df_v['Trækhjul'].fillna("Forhjulstræk") #All Nan to "Forhjulstræk"
df_v['Farve'] = df_v['Farve'].fillna("Other") #All Nan farve to other. Seems to find no pattern in why the color are missing. Only 650 of 9500 are missing, so creating a new catergory
df_v['Price'] = df_v['Price'].str.replace('kr.', '', regex=False).str.strip() #Remove 'Kr.' From the price
df_v['Rækkevidde'] = df_v['Rækkevidde'].str.replace('km', '', regex=False).str.strip() #Remove KM
df_v['Rækkevidde'] = df_v['Rækkevidde'].str.replace('(WLTP) ', '', regex=False).str.strip() #Remove WLTP
df_v['Rækkevidde'] = df_v['Rækkevidde'].str.replace('(NEDC) ', '', regex=False).str.strip() #Remove NEDC
df_v['Batterikapacitet'] = df_v['Batterikapacitet'].str.replace('kWh', '', regex=False).str.strip() #kWh
df_v['Acceleration'] = df_v['Acceleration'].str.replace('sek.', '', regex=False).str.strip() #sek.
df_v['Acceleration'] = df_v['Acceleration'].str.replace(',', '.', regex=False).str.strip() #, to .
df_v['Ydelse'] = df_v['Ydelse'].str.extract(r'(.+hk)', expand=False) #Removes all after 'hk'
df_v['Ydelse'] = df_v['Ydelse'].str.replace('hk', '', regex=False).str.strip() #removes hk
df_v['Price'] = df_v['Price'].str.replace('.', '', regex=False).str.strip() # dots in the price
df_v['Kilometertal'] = df_v['Kilometertal'].str.replace('.', '', regex=False).str.strip() # dots in the kilometertal (Also ran with space)
df_v['Modelår'] = df_v['Modelår'].str.replace('.0', '', regex=False).str.strip() # .0 in modelår
df_v['Batterikapacitet'] = df_v['Batterikapacitet'].str.replace(',', '.', regex=False).str.strip() # replace , with .
```

```python
# Sum up the number of missing values for each column
df_v.isnull().sum()
```

```python
df_v[df_v['Type'].isnull()] #Checking the 3 missing values in Type
```

```python
df_v = df_v.drop(index=5306) #Dropping a row with many missing colums
df_v['Type'] = df_v['Type'].fillna("Cabriolet") #The only 2 missing values on 'type' are Tesla Roadsters, which are considered cabriolet
```

```python
from sklearn.impute import KNNImputer

# List of columns to impute
columns_to_impute = ['Rækkevidde', 'Batterikapacitet', 'Ydelse', 'Acceleration']

# Initialize and apply the KNNImputer
df_v[columns_to_impute] = KNNImputer(n_neighbors=5).fit_transform(df_v[columns_to_impute])
```

```python
df_v
```

```python
df_v['Seller Type'] = df_v['Seller Type'].map({'dealer': True, 'private': False})
df_v['Trækvægt'] = df_v['Trækvægt'].map({'Yes': True, 'No': False})

new_dtypes = {
    'Link': str,
    'Type': 'category',       #SUV, Sedan, hatchback etc.
    'Brand': 'category',      #VW, Tesla, BMW, Audi etc.
    'Model': 'category',      #id.4, Model-3, E-tron etc. 
    'Price': int,             # 'Price' is numric in whole numbers
    'Modelår': int,           # Year as an integer
    'Kilometertal': int,      # Numeric count
    'Rækkevidde': float,      # Numeric range
    'Batterikapacitet': float,# Numeric capacity
    'Ydelse': int,            # Number of Horsepower
    'Acceleration': float,    # Numeric, sec. to 100km/h
    'Farve': 'category',      # Red, blue, black, white etc.
    'Trækhjul': 'category'      #FWD, RWD or AWD
}

# Convert each column to the proper type using the specified data types
# df_v = df_v.astype(new_dtypes)
```

```python
df_v = df_v.astype(new_dtypes)
```
